Merge entre ISCO GROUPS y OCCUPATIONS

```python
isco_occupations = pd.merge(occupations, isco_groups, on='iscoGroup', how='left')

isco_occupations.head(50)
```

```python
from rapidfuzz import process, fuzz

choices = isco_occupations['preferredLabel_y'].tolist()
ai_impact['match_name'] = ai_impact['Job_Title'].apply(lambda x: process.extractOne(x, choices)[0])
ai_impact['match_score'] = ai_impact['Job_Title'].apply(lambda x: process.extractOne(x, choices)[1])
```

```python
choices = isco_occupations['preferredLabel_y'].tolist()
threshold = 90  # solo coincidencias con score >= 90

def match_high(name, choices, threshold=90):
    # Extrae la mejor coincidencia
    match = process.extractOne(name, choices)
    # Si la puntuación es alta, devuelve el nombre; si no, None
    if match and match[1] >= threshold:
        return match[0]
    else:
        return None

# Aplicar
ai_impact['match'] = ai_impact['Job_Title'].apply(lambda x: match_high(x, choices, threshold))
```
